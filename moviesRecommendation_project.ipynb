```python
tags['tag']=tags['tag']+' '
tags.head()
```

```python
agg_func={'tag':'sum'}
tags_a=tags.groupby(tags['movieId']).aggregate(agg_func)
tags_a.head()
```

```python
tags_a=tags_a.reset_index()
tags_a.head()
```

```python
tags_a.shape
```

```python
movies.shape
```

```python
temp=movies.merge(tags_a,on='movieId',how='left')
```

```python
temp['genres']=temp['genres'].apply(lambda x: x.replace('|',' '))
temp
```

```python
temp['tags']=temp['genres']+temp['tag']
temp
```

```python
df2=temp[['movieId','title','tags']]
df2
```

```python
def remove_nan(o):
    
    if pd.isna(o):
        return ''
    else:
        return o
```

```python
df2['tags']=df2['tags'].apply(remove_nan)
```

```python
df2
```

```python
#now making content based
```

```python
#convert tags to lower case
df2['tags']=df2['tags'].apply(lambda x:x.lower())
```

```python
#stemming
import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stemm(text):
    o=list()
    for i in text.split():
        o.append(ps.stem(i))
    return " ".join(o)
```

```python
df2['tags']=df2['tags'].apply(stemm)
df2.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=3000,stop_words='english')
```

```python
vectors=cv.fit_transform(df2['tags']).toarray()
```

```python
vectors
```

```python
vectors.shape
```

```python
#we use cosine similarity
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
similarity.shape
```

```python
#now we recommend top 5 high similarity score movie
sorted(list(enumerate(similarity[1055])),key=lambda x:x[1],reverse=True)[1:6]  # in key we define function that tell what to use for sorting, we can create out own functn also
```

```python
def recommend(movie):
    movie_index=df2[df2['title']==movie].index[0]
    distance=similarity[movie_index]
    movie_list=sorted(list(enumerate(distance)),key=lambda x:x[1],reverse=True)[1:6]

    for i in movie_list:
        print(df2.iloc[i[0]].title)
```

```python
recommend('Thirteen (2003)')
```

```python
#choes from these movies
t=df2[df2['tags']!='']
display(t['title'].to_markdown())
```

```python
recommend('Dead Man Walking (1995)')
```

```python
user_ratings=df1.pivot_table(index='userId',columns='title',values='rating')
user_ratings
```

```python
#we must drop those movies that dont have much users(who rated)
#remove thode movies that have less than 20 users

user_ratings=user_ratings.dropna(thresh=20,axis=1)
print(user_ratings.shape)
```

```python
#fillinginplace of nan
user_ratings=user_ratings.fillna(0)
user_ratings
```

```python
def standardize(row):
    new_row=(row-row.mean())/(row.max()-row.min())
    return new_row
```

```python
user_ratings=user_ratings.apply(standardize)
user_ratings
```

```python
movie_similarity=user_ratings.corr(method='pearson')
movie_similarity
```

```python
def getSimilarMovies(name,rating):
    score=movie_similarity[name]*(rating-2.5)
    score=score.sort_values(ascending=False)
    
    return score[1:6]
```

```python
print(getSimilarMovies('Jumanji (1995)',5))
```

```python
action_lover = [("Avengers, The (2012)",5),("Big Daddy (1999)",1),("Avatar (2009)",5),("Girl, Interrupted (1999)",1)]
similar_scores = pd.DataFrame()
for movie,rating in action_lover:
    similar_scores = similar_scores.append(getSimilarMovies(movie,rating),ignore_index = True)

similar_scores
```
